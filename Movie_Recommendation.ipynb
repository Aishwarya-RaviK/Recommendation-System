```python
import ast
def format(obj):
  list=[]
  for i in ast.literal_eval(obj): 
    list.append(i['name'])
  return list
```

```python
movies_data['genres']=movies_data['genres'].apply(format)
```

```python
movies_data['keywords']=movies_data['keywords'].apply(format)
```

```python
def format_1(obj):
  list=[]
  count=0
  for i in ast.literal_eval(obj):
    if count!=3: 
      list.append(i['name'])
      count=count+1
  return list
```

```python
movies_data['cast']=movies_data['cast'].apply(format_1)
```

```python
def format_2(obj):
  list=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director': 
      list.append(i['name'])
      break
  return list
```

```python
movies_data['crew']=movies_data['crew'].apply(format_2)
```

```python
movies_data.head()
```

```python
movies_data['overview']=movies_data['overview'].apply(lambda x:x.split())
```

```python
movies_data.head()
```

```python
movies_data['genres']=movies_data['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies_data['keywords']=movies_data['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies_data['cast']=movies_data['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies_data['crew']=movies_data['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies_data.head()
```

```python
movies_data['tags']=movies_data['overview']+movies_data['genres']+movies_data['keywords']+movies_data['cast']+movies_data['crew']
```

```python
movies_data.head()
```

```python
data=movies_data[['movie_id','title','tags']]
```

```python
data['tags']=data['tags'].apply(lambda x:" ".join(x))
```

```python
data
```

```python
data['tags']=data['tags'].apply(lambda x: x.lower())
data['title']=data['title'].apply(lambda x:x.lower())
```

```python
data.head()
```

**Stemming** 

```python
import nltk 
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
data['tags']=data['tags'].apply(stem)
```

Vectorization -----> Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vect=cv.fit_transform(data['tags']).toarray()
```

```python
data['tags']=data['tags'].apply(stem)
```

```python
data.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
sim=cosine_similarity(vect)
```

```python
def recommend(movie):
    index = data[data['title'] == movie].index[0]
    distances = sorted(list(enumerate(sim[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(data.iloc[i[0]].title)
  
```

```python
recommend('batman begins')
```
